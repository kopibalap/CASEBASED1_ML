```python
# men DROP tabel Detection_Risk karena tidak berpengaruh 
df.drop(['Detection_Risk'],axis=1)
```

```python

df['LOCATION_ID'].unique()
```

```python
# Menghapus data yang bertipe string
df = df[(df.LOCATION_ID != 'LOHARU')]
df = df[(df.LOCATION_ID != 'NUH')]
df = df[(df.LOCATION_ID != 'SAFIDON')]
df = df.astype(float)
```

```python

df['LOCATION_ID'].unique()
```

## Training


```python
# import library untuk melakukan training
from sklearn.model_selection import train_test_split

# Import library MLPClassifer 
from sklearn.neural_network import MLPClassifier

# Split isi dataset ke variabel baru yaitu X dan Y
X=df[['Sector_score', 'LOCATION_ID', 'PARA_A', 'Score_A', 
      'Risk_A', 'PARA_B', 'Score_B', 'Risk_B', 'numbers', 
      'Score_B.1', 'Risk_C', 'Money_Value', 'Score_MV', 
      'Risk_D', 'District_Loss', 'PROB', 'RiSk_E', 'History', 
      'Prob', 'Risk_F', 'Score', 'Inherent_Risk', 'CONTROL_RISK']]

Y=df['Risk']

# Split dataset ke training set dan test set ( 60 Train : 40 Test )
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=42)


# Membuat objek model
klasifikasi_model = MLPClassifier(hidden_layer_sizes=(6,5),
                    random_state=5,
                    verbose=True,
                    learning_rate_init=0.01,
                    max_iter = 200)

# clasifier data fit ke objek model
klasifikasi_model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Melakukan Prediksi
Ypred=klasifikasi_model.predict(X_test)

# Menghitung nilai  skor akurasi
Skor_akurasi = accuracy_score(Y_test,Ypred) * 1
Skor_presisi = precision_score(Y_test,Ypred) * 1
Skor_recall = recall_score(Y_test,Ypred) * 1

print(" Skor Akurasi yang didapat", Skor_akurasi)
print(" Skor Presisi yang didapat", Skor_presisi)
print(" Skor Recall yang didapat", Skor_recall)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(Y_test,Ypred))
```
